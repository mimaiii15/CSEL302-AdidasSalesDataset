```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("01_Adidas Sales Analysis.csv")

df['Total Sales'] = df.apply(lambda row: row['Units Sold'] * row['Price per Unit'], axis=1)


store_sales = df.groupby('Retailer')['Total Sales'].sum().reset_index()

def categorize_sales_volume(sales):
    """Categorizes stores based on their total sales volume"""
    if sales < 10000:
        return "Low"
    elif sales < 50000:
        return "Medium"
    else:
        return "High"

store_sales['Sales Category'] = store_sales['Total Sales'].apply(categorize_sales_volume)

df = df.merge(store_sales[['Retailer', 'Sales Category']], on='Retailer', how='left')

print(df.head())
```

```python
df
```

```python
high_performing_stores = store_sales[store_sales['Sales Category'] == 'High']

plt.figure(figsize=(10, 6))
df['Sales Category'].value_counts().plot(kind='bar')
plt.title('Distribution of Stores by Sales Volume Category')
plt.xlabel('Sales Volume Category')
plt.ylabel('Number of Stores')
plt.show()

```

```python
top_10_stores = high_performing_stores.nlargest(10, 'Total Sales')
plt.figure(figsize=(12, 8))
plt.barh(top_10_stores['Retailer'], top_10_stores['Total Sales'])
plt.xlabel('Total Sales')
plt.ylabel('Retailer')
plt.title('Top 10 High-Performing Stores')
plt.gca().invert_yaxis()
plt.show()
```
